```python
# ライブラリのインポート
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

# グローバル変数
CARD_NUM = 4
CLASS_NUM = 3
```

```python
# imgフォルダ内の全画像を取得
img_dirs = ['img']
img_data = []
for i, d in enumerate(img_dirs):
    # ./img/以下の各ディレクトリ内のファイル名取得
    files = os.listdir('./'+d)
    for f in files:
        img = cv2.imread('./' + d + '/' + f)
        img_data.append(img)
```

```python
# 画像の表示(1枚選ぶ)
plt.imshow(cv2.cvtColor(img_data[0], cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 1枚ずつ切り取り、対象範囲のRGB値の平均値を取得
def convertImgToRGB(img):
    for card in range(CARD_NUM):
        trim_img = img[1180:1200,(80 + 113 * card):(193 + 113 * card)]
        plt.imshow(cv2.cvtColor(trim_img, cv2.COLOR_BGR2RGB))
        plt.show()
        averages = trim_img.mean(0).mean(0)
        img_data.append(averages)

convertImgToRGB(img_data[0])
```
